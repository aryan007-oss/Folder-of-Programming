# Convert Celcius to Fahrenheit and Kelvin

```python
import temperature as temp
c = float(input("Celcius: "))

print(f"{c}° Celcius = {temp.celcius_to_fahrenheit(c)} Fahrenheit")
print(f"{c}° Celcius = {temp.celcius_to_kelvin(c):.2f} Kelvin")
```

# Convert Kelvin to Celcius, Fahrenihet 

```python
import temperature as temp
k = float(input("Kelvin: "))


print(f"{k} Kelvin = {temp.kelvin_to_fahrenheit(k)} Fahrenheit")
print(f"{k} Kelvin = {temp.kelvin_to_celcius(k):.2f} Celcius")
```
